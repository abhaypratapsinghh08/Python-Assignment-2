#### Problem Statement 1: Samantha has created a dataset named 'top50spotify.csv' of her top 50 songs from spotify.

1. Import the dataset as a DataFrame and drop the first column.
2. Save it as 'top50.csv'.
3. Find the average Energy and Length of first 10 songs.
4. Find the total length of songs, group by genre from top to bottom.
5. Print the artist name with the most number of tracks in one genre. (Hint: Group by artist name and genre)
6. Print the data of the tracks created by the artist from the previous question.

### 1.Import the dataset as a DataFrame and drop the first column.

```python
import pandas as pd
import numpy as np
```

```python
spotify = pd.read_csv(r"C:\\Python\Downloads\top50spotify.csv", header=0)
spotify = spotify.drop(spotify.columns[0], axis=1)
spotify
```

### 2.Save it as 'top50.csv'

```python
spotify.to_csv("top50.csv")
```

### 3.Find the average Energy and Length of first 10 songs.

```python
spotify_1=spotify.head(10)
spotify_2 = spotify_1[['Energy', 'Length.']].mean()
spotify_2
```

### 4.the total length of songs, group by genre from top to bottom.

```python
spotify.groupby('Genre')['Length.'].sum().sort_values(ascending=False)
```

### 5.the artist name with the most number of tracks in one genre.

```python
spotify_3 = spotify.groupby(['Genre','Artist.Name'])['Track.Name'].count()
spotify_3[spotify_3 == spotify_3.max()]
```

### 6.the data of the tracks created by the artist from the previous question.

```python
spotify_3 = spotify.groupby(['Genre','Artist.Name'])['Track.Name'].count()
spotify_3
```

# Problem Statement 2: Write a Python program to perform the following tasks.

1. Create a pandas series from the below dictionary where indices are subjects.
2. Convert the created series into DataFrame and replace the null values with zeroes.
3. Transpose the DataFrame and create a new column 'Average' and fill the values in it by calculating the average of all subjects.

### 1.Create a pandas series from the below dictionary where indices are subjects:

```python
import pandas as pd
import numpy as np
```

```python
print('Creating a series from the dictionary')
dictionary = {'English':{'Sam':60,'Jackson':74,'Ahree':85} , 
              'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},
              'Geography':{'Jackson':92, 'Gloria':95, 'Isla':82, 'Aron':75, 'Ahree':76},
              'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},
              'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}}
ser = pd.Series(dictionary)
ser
```

### 2.Convert the created series into DataFrame and replace the null values with zeroes.

```python
print('Creating a data frame and replacing the nun with 0')
df = ser.apply(pd.Series)
df1 = df.fillna(0)
df1
```

### 3.Transpose the DataFrame and create a new column 'Average' and fill the values in it by calculating the average of all subjects

```python
df2 = df1.transpose()  # transpose the data
df2
```

```python
df3 = df2.mean(axis=1)   # Calculating the mean value
df2['Average'] = df3     # Adding the new column as Average and putting the mean value in it
df2
```

# Problem Statement 3: Write a Python program to create a series from 1 to 1000 and select only numbers divisible by 7 and 17.

### Create a series from 1 to 1000 and select only numbers divisible by 7 and 17.

```python
import pandas as pd
import numpy as np
```

```python
ab =[]
for x in range(1, 1000):
    if (x%7==0) and (x%17==0):
        ab.append(str(x))
print('Numbers divisible by 7 and 17 :')
print(ab)
```

```python
print("series of numbers which are divisible by 7 and 17 :")
df = pd.Series(ab)
df
```
